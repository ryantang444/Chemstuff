```python
def averages(df: pd.DataFrame):
    ranking = {}
    for i in range(37):
        mean = (df.iloc[i,2] + df.iloc[i,3] + df.iloc[i,4]) / 3
        ranking[df.iloc[i,0] + ' ' + df.iloc [i,1]] = mean
    
    sortedranking = sorted(ranking.items(), key=lambda x:x[1], reverse=True)
    return sortedranking

sortedranking = averages(df)


# make sure they do not end with a space
```

```python
a_list = ['BIO 105','CHEM 120','PHYS 110','MATH 110']

def combine_pairs(my_list: list[str]):
    a = list(itertools.combinations(my_list, 2))
    y = [' '.join(i) for i in a]
    return y

def alphabetize_pairs(pair: str):
    a = pair.find(" ") + 1
    space = pair.find(" ", a)
    if pair[0] > pair[space + 1]:
        pair = pair[space + 1:] + ' ' + pair[0:space]
        return pair
        
def alphabetize_list(my_list: list[str]):
    final = []
    for i in range(len(my_list)):
        space_1 = my_list[i].find(" ")
        space_2 = my_list[i].find(" ", space_1 + 1)
        if my_list[i][0] > my_list[i][space_2 + 1]:
            final.append(alphabetize_pairs(my_list[i]))
        else:
            final.append(my_list[i])
    return final

def check_pairs(my_list: list[str]):
    total_coenrolls = 0
    for i in range(len(sortedranking)):
        for j in range(len(my_list)):
            if sortedranking[i][0] == my_list[j]:
                total_coenrolls += sortedranking[i][1]
                print(sortedranking[i][0] + " ("  + str(sortedranking[i][1]) + ")")
    return total_coenrolls

def list_to_score(my_list: list[str]):
    return check_pairs(alphabetize_list(combine_pairs(a_list)))
list_to_score(a_list)
```

```python
def courseDictGen(totaldf):
    courseCounts = {}
    courseCounts['Courses'] = []
    for i in range(totaldf.shape[0]):
        newDict = {}
        newDict['Name'] = totaldf.iloc[i, 3] + ' ' + totaldf.iloc[i, 4]
        newDict['Course Type'] = totaldf.iloc[i,6]
        newDict['Days'] = totaldf.iloc[i,9]
        newDict['Time'] = totaldf.iloc[i,10] + "-" + totaldf.iloc[i,11]
        courseCounts['Courses'].append(newDict)
    return courseCounts


def ScheduleGenerate(CourseDict):
    Schedule = {}
    for i in range(len(CourseDict['Courses'])):
        if CourseDict['Courses'][i]['Course Type'] == 'Lecture':
            a = CourseDict['Courses'][i]['Time']
            b = CourseDict['Courses'][i]['Name']
            c = CourseDict['Courses'][i]['Days']
            if a + c in Schedule:
                Schedule[a + c].append(b)
            else:
                Schedule[a + c] = [b]
    return Schedule

ScheduleGenerate(courseDictGen(fall_2022))

#print(totaldf.iloc[0,6])
```
