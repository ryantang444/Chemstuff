```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
```

```python
df = pd.read_csv("/content/Fall Enrollment Data - Sheet1 (1).csv")
df
```

```python
ranking = {}

sortedranking = 37
for i in range(sortedranking):
    mean = (df.iloc[i,3] + df.iloc[i,4] + df.iloc[i,5]) / 3
    ranking[df.iloc[i,1] + ' ' + df.iloc [i,2]] = mean

sortedranking = sorted(ranking.items(), key=lambda x:x[1], reverse=True)
print(sortedranking)
```

```python
# MOST RECENT
a_list = ['BIO 105','CHEM 120','PHYS 110','MATH 110']


def combine_pairs(my_list: list[str]):
    a = list(itertools.combinations(my_list, 2))
    y = [' '.join(i) for i in a]
    return y

def alphabetize_pairs(pair: str):
    a = pair.find(" ") + 1
    space = pair.find(" ", a)
    if pair[0] > pair[space + 1]:
        pair = pair[space + 1:] + ' ' + pair[0:space]
        return pair

def alphabetize_list(my_list: list[str]):
    final = []
    for i in range(len(my_list)):
        a = my_list[i].find(" ") + 1
        space = my_list[i].find(" ", a)
        if my_list[i][0] > my_list[i][space + 1]:
            final.append(alphabetize_pairs(my_list[i]))
        else:
            final.append(my_list[i])
    return final


def check_pairs(my_list: list[str]):
    total_coenrolls = 0
    for i in range(len(sortedranking)):
        for j in range(len(my_list)):
            if sortedranking[i][0] == my_list[j]:
                total_coenrolls += sortedranking[i][1]
                print(sortedranking[i][0], "have a co-enrollment score of", sortedranking[i][1])
    return total_coenrolls

def list_to_score(my_list: list[str]):
    return check_pairs(alphabetize_list(combine_pairs(a_list)))

print("Total mean value of schedule:", (a_list), ":", list_to_score(a_list))
print("A higher co-enrollment score means that this set of classes is more commonly taken together.")


```
