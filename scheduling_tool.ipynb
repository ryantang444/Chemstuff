```python
def averages(falldf: pd.DataFrame):
    ranking = {}
    for i in range(37):
        mean = (falldf.iloc[i,2] + falldf.iloc[i,3] + falldf.iloc[i,4]) / 3
        ranking[falldf.iloc[i,0] + ' ' + falldf.iloc [i,1]] = mean

    fallsortedranking = sorted(ranking.items(), key=lambda x:x[1], reverse=True)
    return fallsortedranking

fallsortedranking = averages(falldf)
print(fallsortedranking)


# make sure they do not end with a space
```

```python
def averages(springdf: pd.DataFrame):
    ranking = {}
    for i in range(28):
        mean = (springdf.iloc[i,2] + springdf.iloc[i,3] + springdf.iloc[i,4]) / 3
        ranking[springdf.iloc[i,0] + ' ' + springdf.iloc [i,1]] = mean

    springsortedranking = sorted(ranking.items(), key=lambda x:x[1], reverse=True)
    return springsortedranking

springsortedranking = averages(springdf)
print(springsortedranking)


# make sure they do not end with a space
```

```python
fall_list = ['BIO 105','CHEM 120','PHYS 110','MATH 110']
spring_list = ['BIO 130', 'CHEM 120', 'BIO 110', 'MATH 110']


def combine_pairs(my_list: list[str]):
    a = list(itertools.combinations(my_list, 2))
    y = [' '.join(i) for i in a]
    return y

def alphabetize_pairs(pair: str):
    a = pair.find(" ") + 1
    space = pair.find(" ", a)
    if pair[0] > pair[space + 1]:
        pair = pair[space + 1:] + ' ' + pair[0:space]
        return pair

def alphabetize_list(my_list: list[str]):
    final = []
    for i in range(len(my_list)):
        space_1 = my_list[i].find(" ")
        space_2 = my_list[i].find(" ", space_1 + 1)
        if my_list[i][0] > my_list[i][space_2 + 1]:
            final.append(alphabetize_pairs(my_list[i]))
        else:
            final.append(my_list[i])
    return final

def fall_check_pairs(my_list: list[str]):
    total_coenrolls = 0
    for i in range(len(fallsortedranking)):
        for j in range(len(my_list)):
            if fallsortedranking[i][0] == my_list[j]:
                total_coenrolls += fallsortedranking[i][1]
                print(fallsortedranking[i][0] + " ("  + str(fallsortedranking[i][1]) + ")")
    return total_coenrolls

def spring_check_pairs(my_list: list[str]):
    total_coenrolls = 0
    for i in range(len(springsortedranking)):
        for j in range(len(my_list)):
            if springsortedranking[i][0] == my_list[j]:
                total_coenrolls += springsortedranking[i][1]
                print(springsortedranking[i][0] + " ("  + str(springsortedranking[i][1]) + ")")
    return total_coenrolls

def fall_list_to_score(my_list: list[str]):
    return fall_check_pairs(alphabetize_list(combine_pairs(fall_list)))

def spring_list_to_score(my_list: list[str]):
    return spring_check_pairs(alphabetize_list(combine_pairs(spring_list)))

print("Fall")
fall_list_to_score(fall_list)
print("Spring")
spring_list_to_score(spring_list)

```
